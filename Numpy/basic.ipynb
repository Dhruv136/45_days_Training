NumPy (Numerical Python) is a powerful Python library used for numerical computing. It provides support for:

Multi-dimensional arrays (ndarray)
Mathematical operations on arrays
Linear algebra
Statistical functions
Random number generation
Fast execution compared to Python lists

```python
import numpy as np

arr1= np.array([1,2,3,4])
print(arr1)
```

```python
print(np.__version__)
```

```python
# 2d arrays
arr2 = np.array([[1,2],[3,4]])
print(arr2)
```

```python
lst = [1,2,3,4,5]
print("this is my list", lst)
print("type of my list",type(lst))
print("list is converted into array")
arr=np.array(lst)
print("this is array", arr)
print("type of my array", type(arr))
```

```python
arr = np.zeros(50)
print(arr)
```

```python
arr = np.ones(20)
print(arr)
```

```python
arr= np.zeros([5,10])
arr
```

```python
arr = np.arange(60)
print(arr)
arr=arr.reshape(12,5) # converting of 1d array into 2d array
print(arr)
```

```python
# array attributes or function 
arr = np.array([[1,3],[5,3]])
print(arr)
print(arr.shape)  
print(arr.ndim)    
print(arr.size)    
print(arr.dtype)
```

```python
#slicing in fucking array
arr[0:5]
```
