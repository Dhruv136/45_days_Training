Machine Learning (ML)

1. Definition

1.1 Machine Learning is a branch of Artificial Intelligence (AI).

1.2 It enables computers to learn from data without explicit programming.

 

2. Types of Machine Learning

2.1 Supervised Learning – Learns from labeled data.

2.2 Unsupervised Learning – Finds patterns in unlabeled data.

2.3 Reinforcement Learning – Learns using rewards and penalties
.

 

3. Key Terms
3.1 Dataset – Collection of data.

3.2 Features – Input variables.

3.3 Label/Target – Output variable.

3.4 Training – Process of learning from data.

3.5 Testing – Evaluation of model performance.

3.6 Model – Algorithm that learns patterns.

 

4. Common Algorithms
4.1 Linear Regression.

4.2 Logistic Regression.

4.3 Decision Tree.

4.4 Random Forest.

4.5 K-Nearest Neighbors (KNN).

4.6 Support Vector Machine (SVM).

 

5. Applications

5.1 Recommendation Systems.

5.2 Spam Detection.

5.3 Image Recognition.

5.4 Medical Diagnosis.

5.5 Fraud Detection.

 

6. Conclusion
6.1 Machine Learning helps computers learn from data and make predictions automatically.

6.2 It is widely used in modern technology and automation.
