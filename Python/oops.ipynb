```python
#  singlen inheritamce
class company:
    pass
class Developer(company):
    pass
```

```python
class Company:
    company_name="dhruv di company"
    company_add="jaipur"
class Developer(Company):
    developer_name="buhua"
dev=Developer()
```

```python
#multiple inheritance
class Company:
    company_name="dhruv di company"
    company_add="jaipur"
class Client:
    client_name="huha"

class Developer(Company,Client):
    developer_name="dhruv"
dev=Developer()
print(dev.company_name)
print(dev.developer_name)    
print(dev.client_name)    
```

```python
#multilevel inheritance

class Grandfather:
    grand_father="murarilal"
class Father(Grandfather):
    father_name="manoj kokcha"
class Chinu(Father):
    name="chinu da boss"
ch=Chinu()
print(ch.grand_father)
print(ch.father_name)
print(ch.name)

```

```python
#Hierarchial inheritance
class Parent:
    car = "BMW"

class Son(Parent):
    pass

class Daughter(Parent):
    pass

print(Son.car)
print(Daughter.car)
```

```python
#Hybird inheritance
class Grandparent:
    land = 50

class Father(Grandparent):
    pass

class Mother(Grandparent):
    pass

class Child(Father, Mother):
    pass

c = Child()

print(c.land)
```

```python
# Encapsulation in python

class Mompany:
    company_name="dhruv di company"# public variable
    _company_add="jaiuuu"# protected variable
    __company_acc="749347903" #private variable
    def show1(self):
        print(self.__company_acc) #accessing private variable
        
hu=Mompany()  #accessing private
hu.show1()#accessing private

```
