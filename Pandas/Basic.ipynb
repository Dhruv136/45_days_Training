Pandas (Theory)

Pandas is an open-source Python library used for data analysis and data manipulation. It helps in handling and organizing large amounts of data in a structured form.

Features of Pandas
Fast and efficient data handling.
Easy reading and writing of data from files (CSV, Excel, etc.).
Supports data cleaning and preprocessing.
Provides statistical analysis tools.
Handles missing data effectively.
Main Data Structures
1. Series
A one-dimensional labeled array.
Can store data of any type (integers, strings, floats, etc.).
Similar to a single column in a table.
2. DataFrame
A two-dimensional data structure.
Consists of rows and columns.
Similar to a spreadsheet or database table.
Most commonly used structure in Pandas.
Common Uses of Pandas
Data cleaning
Data transformation
Data analysis
Data visualization preparation
Working with CSV and Excel files
Advantages
Easy to learn and use.
Saves time when working with large datasets.
Integrates well with libraries like NumPy and Matplotlib.

```python
import numpy as np 
data = np.array([["a",25, 5000],["b",23,400]])
print(data)
```

```python
import pandas as pd
df = pd.DataFrame([["a",25,50000]
                   ,["b",23,40000]])
print(df)
```

```python
data = {
    'name':['alice','bob','charlie'],
    'Age':[23,45,64],
    'city':['new york','london','paris']
}
df=pd.DataFrame(data)
print(df)
```

```python
s = pd.Series([80,70,60])
print(s)
#index based printing
print(s[0])
```

```python
p = pd.Series([80,40,70],index=["math","science","english"])
print(p)
#label bases printing
print(p["math"])


```

```python
import pandas as pd
d = pd.Series([10,30,30])
print(d+5)
print(d*6)
```

```python
#diff between dataframe and series
data = {
    "name":["puja","rahul","ankit"],
    "Age":[2,3,4],
    "Marks":[50,60,50]
}
df = pd.Series(data)
print(df)
df = pd.DataFrame(data)
print(df)
```

```python
#adding a column
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ishika", "dhruv", "Bagri"],
    "Age": [19, 20,50]
})
print(df)
df["City"] = ["loharu", "ramgarh","pilani"]

print(df)
```

```python
#remove a column 
df = df.drop("City", axis=1)

print(df)
```

```python
#rename a column
df.rename(columns={"Student_Name": "Nigga_name"}, inplace=True)

print(df)
```
